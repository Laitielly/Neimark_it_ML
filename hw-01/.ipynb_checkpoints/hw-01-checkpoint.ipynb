Какое кол-во людей, имеющих х трейнингов, получили повышение:

```python
def distribution_pos_tar(X, y, col):
    X_copy = X.copy()
    X_copy['y'] = y
    b = X_copy.groupby(by=[col])['y'].value_counts()
    print(b)
    print()
    print('Процент людей:')
    array = X_copy[col].unique()
    array.sort()
    for i in array:
        if i not in b: continue
        if b[i].size == 2:
            print(f'i = {i}, % = {(b[i][1] / X_train.shape[0]) * 100}, соотношение тех, кого повысили, к иным - {b[i][1] / b[i][0]}')
        else:
            print(f'i = {i}, % = 0')
```

```python
distribution_pos_tar(X_train, y_train, 'no_of_trainings')
```

Видно, что подавляющая часть прошла лишь 1 трейнинг - 81%. Визуально можно предположить, что кол-во трейнингов не особо влияет на повышение.

#### Выбросы

Думаю нет смысла читать выбросы у признаков, которые принимают значение 0 и 1. Поэтому я считаю выбросы у числовых признаков, где выбросы могут находиться.

```python
num = ['age', 'length_of_service', 'avg_training_score']
```

```python
outliers = 0

for i in num:
    Q1, Q3 = X_train[i].quantile([.25, .75])
    IQR = Q3 - Q1
    price_lower = Q1 - 1.5*IQR
    price_upper = Q3 + 1.5*IQR
    
    outliers += (X_train.shape[0] - ((X_train[i] < price_lower) | (X_train[i] > price_upper)).value_counts()[False]) / X_train.shape[0] * 100
    
outliers
```

Колонка age будет удалена, так что не думаю, что отсюда стоит что-то выбрасывать, ведь возраст мало как по мне влияет на повышение. length_of_service сильно влияет на повышение, как мы поняли из анализа, как и avg_training_score, поэтому выбрасывать их тоже не стои. Чем выше данные значение, тем охотнее таких людей повышают.

### Обработать пропущенные значения (или убедиться, что их нет)
#### Train

```python
cols_null = X_train.columns[X_train.isnull().any()]
missed = pd.DataFrame(X_train[cols_null].isnull().sum().sort_values(ascending=False) / X_train.shape[0], columns=['% NULL'])
missed
```

Так как пропусков немного, то заменим числовые на медиану, а категориальные на часто встречаемые значения.

### Попробуйте как минимум 3 метода классификации (регрессии). Объясните ваш выбор. Найдите значения метрик на обучающей и тестовой выборке. Сделайте вывод

**Я решила попробовать 5 моделей:**
- Логистическая регрессия - базовая и простая модель для классификации. 
- Решающие деревья - более сложная модель, в целом предполагаю, что алгоритмы на основе решающих деревьев дадут результаты выше.
- Extremely Randomized Trees Classifier(Extra Trees Classifier) - ансамбль деревьев, который агрегирует результаты нескольких декоррелированных деревьев решений, собранных в «лесе». Похож на случайный лес.
- Random Forest - в целом алгоритм выше похож на случайный лес, однако я взяла его, чтобы сравнить результаты из интереса.
- xgboost classifier - алгоритм, основанный на дереве поиска решений и использующий фреймворк градиентного бустинга + оптимизация.

**Метрики:**
- balanced_accuracy_score
- accuracy_score
- f1-score
- average_precision_score

```python
def classifier(model, X, y, X_test, y_test):
    model.fit(X, y)
    
    pred_train = model.predict(X)
    pred_test = model.predict(X_test)
    
    print(f'balanced_accuracy_score train - {balanced_accuracy_score(y, pred_train)}\n' \
         f'f1 score train - {f1_score(y, pred_train)}\n' \
         f'average_precision_score train - {average_precision_score(y, pred_train)}\n' \
         f'accuracy_score train - {accuracy_score(y, pred_train)}\n' \
         f'confusion matrix - \n{confusion_matrix(y, pred_train)}')
    
    print()
    print()
          
    print(f'balanced_accuracy_score test - {balanced_accuracy_score(y_test, pred_test)}\n' \
         f'f1 score test - {f1_score(y_test, pred_test)}\n' \
         f'average_precision_score test - {average_precision_score(y_test, pred_test)}\n' \
         f'accuracy_score test - {accuracy_score(y_test, pred_test)}\n' \
         f'confusion matrix - \n{confusion_matrix(y_test, pred_test)}')
```

#### Логистическая регрессия

```python
Log_reg = LogisticRegression()
classifier(Log_reg, X_train_prepared, y_train, X_test_prepared, y_test)
```

#### Решающие деревья

```python
Dec_tree = DecisionTreeClassifier()
classifier(Dec_tree, X_train_prepared, y_train, X_test_prepared, y_test)
```

#### Extremely Randomized Trees Classifier(Extra Trees Classifier)

```python
etc = ExtraTreesClassifier(verbose=False)
classifier(etc, X_train_prepared, y_train, X_test_prepared, y_test)
```

#### Random Forest

```python
rfc = RandomForestClassifier()
classifier(rfc, X_train_prepared, y_train, X_test_prepared, y_test)
```

#### xgboost classifier

```python
xgb = XGBClassifier()
classifier(xgb, X_train_prepared, y_train, X_test_prepared, y_test)
```

Такое ощущение, что все деревья переобучились. По тестовому значению выбились - XGBClassifier, DecisionTreeClassifier. Я решила попробовать еще Catboost.

Пелена кода - поиск оптимального итератора, с учетом установленного precision. Разбираться в коде не обязательно) Но кратко - инициализируем массивы полученных оценок, а именно - pr, итерации, точность(баланс), threshold для лучшей точности и для лучшего pr.

Потом по списку перебираем итерации. Делаем кросс-валидацию, потом для трейна балансировку, обучаем, сохраняем (так для каждой части кросс-валидации), выбираем threshold, считаем метрики, усредняем эти оценки и проверяем является ли данная оценка наибольшей. 


Последняя строка показывает лучшие на обучающей. Однако обирать будем по графикам с учетом валидации.

```python
iterations = [i for i in range (100, 501, 20)]
    
cross_val = KFold(n_splits=3, shuffle=True)

best_pr = -1
best_iter_true = -1

best_acc = -1
best_iter_acc = -1

all_pr_train, all_pr_val = [], []
all_acc_train, all_acc_val = [], []

threshold_all = []
best_threshold_pr = -1
best_threshold_acc = -1
    
for iterat in iterations:
    pr_now_train = []
    pr_now_val = []
    
    acc_now_train = []
    acc_now_test = []
    
    threshold_now = []
    
    for X_ind_train, X_ind_test in cross_val.split(X_train_prepared):
        X_train_val, X_test_val, y_train_val, y_test_val = X_train_prepared.iloc[X_ind_train, :], \
        X_train_prepared.iloc[X_ind_test, :], y_train.iloc[X_ind_train, :], y_train.iloc[X_ind_test, :]
        
        model = CatBoostClassifier(verbose=False, iterations=iterat, \
                                     learning_rate=0.1, depth=10)
        
        X_res, y_res = ADASYN().fit_resample(X_train_val, y_train_val)
        
        model.fit(X_res, y_res)

        y_pred_val = model.predict_proba(X_test_val)[:,1]
        y_pred_train = model.predict_proba(X_train_val)[:,1]
        
        ########
        
#         avg_pr_train = average_precision_score(y_train_val, y_pred_train)
#         avg_pr_test = average_precision_score(y_test_val, y_pred_val)
        
        precision, recall, thresholds = precision_recall_curve(y_train_val, y_pred_train)
        
        best_threshold = 0
        max_recall = 0
        
        for i, x in enumerate(precision):
            if x >= 0.6 and recall[i] >= max_recall:
                best_threshold = thresholds[i]
                max_recall = recall[i]
                
        y_pred_train = y_pred_train >= best_threshold
        y_pred_val = y_pred_val >= best_threshold
        
        threshold_now.append(best_threshold)
        
#         print(f'Итерация - {iterat}, best_threshold здесь = {best_threshold}. - > ' \
#               f'recall train = {recall_score(y_train_val, y_pred_train)}, recall val = {recall_score(y_test_val, y_pred_val)} -> '\
#               f'precision train = {precision_score(y_train_val, y_pred_train)}, precision val = {precision_score(y_test_val, y_pred_val)}')
 
    
        avg_pr_train = average_precision_score(y_train_val, y_pred_train)
        avg_pr_val = average_precision_score(y_test_val, y_pred_val)
        
        pr_now_train.append(avg_pr_train)
        pr_now_val.append(avg_pr_val)
        
        #########
        
        avg_acc_train = balanced_accuracy_score(y_train_val, y_pred_train)
        avg_acc_test = balanced_accuracy_score(y_test_val, y_pred_val)
        
        acc_now_train.append(avg_acc_train)
        acc_now_test.append(avg_acc_test)
    
    ########
    
    all_pr_train.append(np.mean(pr_now_train))
    all_pr_val.append(np.mean(pr_now_val))
    
    ######
    
    all_acc_train.append(np.mean(acc_now_train))
    all_acc_val.append(np.mean(acc_now_test))
    
    threshold_all.append(np.mean(threshold_now))
    
    if (np.mean(pr_now_train) > best_pr):
        best_iter_true = iterat
        best_pr = np.mean(pr_now_train)
        best_threshold_pr = np.mean(threshold_now)
      
    
    if (np.mean(all_acc_train) > best_acc):
        best_iter_acc = iterat
        best_acc = np.mean(all_acc_train)
        best_threshold_acc = np.mean(threshold_now)
            
    print(f'Итерация - {iterat} -> pr now - {"{:.11}".format(np.mean(pr_now_train))}' \
          f' -> баланс точность - {"{:.11}".format(np.mean(all_acc_train))} ->' \
          f' threshold = {"{:.11}".format(np.mean(threshold_now))}')
            
print(f'\nЛучший pr - {best_pr} при кол-ве итераций = {best_iter_true} и threshold = {best_threshold_pr}\n'\
f'Лучший acc - {best_acc} при кол-ве итераций = {best_iter_acc} и threshold = {best_threshold_acc}')
```

```python
def con_mat(cf_matrix):
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in
                    cf_matrix.flatten()]
    group_percentages = ["{0:.2%}".format(value) for value in
                         cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
              zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
```
